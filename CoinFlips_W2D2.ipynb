```python
import random
import math
import statistics
import matplotlib.pyplot as plt

%matplotlib inline

def flip_a_coin():
    """Randomly chooses a 1 or 2 and returns it.  1 is heads and 2 is tails."""
    return random.randint(1, 2)


def add_total_coin(coin_dict, head_count, tail_count):
    """Adds the total coins at the increments in the dict/list."""
    coin_dict[1].append(head_count)
    coin_dict[2].append(tail_count)
    return coin_dict


def subtraction_coins(list_of_heads, list_of_tails):
    """Subtracts elements from two lists and returns new list."""
    new_coins_list = []
    count = 0
    while count <= len(list_of_heads) - 1:
        new_coins_list.append(list_of_heads[count] - list_of_tails[count])
        count += 1
    return new_coins_list


def coins_ratio(list_of_heads, list_of_tails):
    """Divides one list by another and returns new list."""
    new_coins_list = []
    count = 0
    
    while count <= len(list_of_heads) - 1:
        if list_of_tails[count] == 0:
            new_coins_list.append(0)
            count += 1
        else:
            new_coins_list.append(list_of_heads[count] / list_of_tails[count])
            count += 1
    return new_coins_list


def heads_to_total_ratio(list_of_heads, list_of_tails):
    """Returns ratio of heads to total flips."""
    new_coins_list = []
    count = 0
    while count <= len(list_of_heads):
        new_coins_list.append(list_of_heads[count] / (list_of_heads[count] + list_of_tails[count]))
        count += 1
    return new_coins_list
        

def run_simulation(n):
    """Runs simulation n times and records the coin flip at 2^n increments.  1 is heads and 2 is tails"""
    random_coin = 0
    count = 0
    dict_of_coins = {1: [], 2: []}
    count_of_heads = 0  # same as 1
    count_of_tails = 0  # same as 2
    inc = n
    floor_inc = (inc // inc) - 1
    heads_minus_tails = []
    ratio_heads_to_tails = []
    
    while (count <= 2**inc + 1):
        random_coin = flip_a_coin()
        if random_coin == 1:
            count_of_heads += 1
        else:
            count_of_tails += 1
            
        if count == 2**floor_inc:
            dict_of_coins = add_total_coin(dict_of_coins, count_of_heads, count_of_tails)
            floor_inc += 1
        count += 1
    print(dict_of_coins)
    return dict_of_coins
    
    
heads_tails_dict = {}
heads_minus_tails = []
ratio_heads_to_tails = []
ratio_heads_to_total_flips = []

heads_tails_dict = run_simulation(100)
heads_minus_tails = subtraction_coins(heads_tails_dict[1], heads_tails_dict[2])
ratio_heads_to_tails = coins_ratio(heads_tails_dict[1], heads_tails_dict[2])
ratio_heads_to_total_flips = heads_to_total_ratio(heads_tails_dict[1], heads_tails_dict[2])

plt.hist(ratio_heads_to_total_flips, bins=25)
plt.xlabel("Heads to Total Flips")
plt.ylabel("Total Flips")
plt.title("Ratio of Heads to Total Flips")

ymin, ymax = plt.ylim()
mean = statistics.mean(ratio_heads_to_total_flips)
stdev = statistics.pstdev(ratio_heads_to_total_flips)
plt.vlines(mean, ymin, ymax, color="blue")
plt.vlines([mean - 2* stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, linestyles="dashed", 
           color="red")
plt.show()
```
